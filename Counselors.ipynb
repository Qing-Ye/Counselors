```python
behavior = ['Bintegral2ed', 'BISPintegral', 'BPPSmajor', 'Bgrade8',
       'Bannualrw', 'Bpexpect', 'Badminsupport', 'Bmotivate', 'Bpost2nd',
       'Bparentcomm', 'Bparentinvolve', 'Bstrengthweak', 'Bcourseselect',
       'Bimproveac', 'Bpost2ndsuc', 'Bstudentknow', 'Bevaluate']

practice_EP = ['EPhelpstudentID', 'EPawaredmskills', 'EPdevrxskills', 'EPacquireknow',
       'EPlearnimpPPS', 'EPmiddleschool', 'EPgraduateoptions', 'EPpostHSplan']

practice_P = ['PallstudPPS', 'Pcanspeakknow', 'PincCPandCC',
       'PstudentscanrwPPS', 'PdistrictpolicyPPS', 'PdistrictCSIP', 'Pprofdev',
       'Pstaffinvolve', 'Pevaluate']
```

```python
from factor_analyzer import FactorAnalyzer
```

```python
df_behavior = df[behavior].dropna(axis= 0)
```

```python
df_behavior.info()
```

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value=calculate_bartlett_sphericity(df_behavior)
chi_square_value, p_value
```

In this Bartlettâ€™s test, the p-value is 0. The test was statistically significant, indicating that the observed correlation matrix is not an identity matrix.

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df_behavior)
kmo_model
```

The overall KMO for our data is 0.94, which is excellent. This value indicates that could do factor analysis.

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer()
fa.analyze(df_behavior, 17, rotation=None)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev
```

```python
# Create scree plot using matplotlib
import matplotlib.pyplot as plt

plt.scatter(range(1,df_behavior.shape[1]+1),ev)
plt.plot(range(1,df_behavior.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

2-factors eigenvalues are greater than one. It means need to choose only 2 factors 

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer()
fa.analyze(df_behavior, 2, rotation='varimax')
fa.loadings
```

Factor 1 has high factor loadings for Bintegral2ed, BISPintegral, BPPSmajor, Bgrade8, Bannualrw, Bpexpect, and Badminsupport.
Factor 2 has high factor loadings for Bmotivate, Bpost2nd, Bparentcomm, Bparentinvolve, Bstrengthweak, Bcourseselect, Bimproveac, Bpost2ndsuc, Bstudentknow, and Bevaluate.

```python
fa.get_factor_variance()
```

Total 55% cumulative variacne explained by the the 2 factors.
